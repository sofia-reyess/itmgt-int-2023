```python
import requests
import time

def generate_data_points(num_points):
    data_points = []

    for _ in range(num_points):
        # Make a request to the API
        response = requests.get("https://api.randomuser.me/")
        data = response.json()

        # Extract latitude, longitude, and timestamp
        latitude = float(data["results"][0]["location"]["coordinates"]["latitude"])
        longitude = float(data["results"][0]["location"]["coordinates"]["longitude"])
        timestamp = int(time.time())

        # Append the data point to the list
        data_points.append({"latitude": latitude, "longitude": longitude, "timestamp": timestamp})

        # Wait for 10 seconds before making the next request
        time.sleep(10)

    return data_points

# Generate at least 100 data points
num_points = 100
data_points = generate_data_points(num_points)

# Print the generated data points
for point in data_points:
    print(point)
```

```python
import matplotlib.pyplot as plt

latitudes = [point["latitude"] for point in data_points]
longitudes = [point["longitude"] for point in data_points]
timestamps = [point["timestamp"] for point in data_points]

plt.figure(figsize=(10, 6))
sc = plt.scatter(longitudes, latitudes, c=timestamps, alpha=0.7, s=50)

plt.colorbar(sc, label='Timestamp')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('ISS Location Through Time')

plt.grid()
plt.show()
```
